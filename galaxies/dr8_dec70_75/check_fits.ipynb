```python
from astropy.table import Table
cat = Table.read('dr8_photoz_dec70_75_full.fits', memmap=True)
cat.info()
```

```python
print(cat.colnames)
print(cat['z_phot_mean'].unit)
```

```python
import matplotlib.pyplot as plt
plt.hist(cat['z_phot_mean'], bins=300, range=(0,1.5), histtype='step')
plt.xlabel('photo-z'); plt.ylabel('count')
plt.title('Legacy Survey DR8 • 70° < δ < 75°')
```
