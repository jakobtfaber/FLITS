# Radon/Hough Transform DM Estimator

This notebook implements and stress-tests the Radon/Hough transform approach to DM estimation.

## Key Idea
Transform the dynamic spectrum to (ν⁻², t) coordinates where dispersion curves become **straight lines**. Then use efficient line detection (Radon/Hough transform) to find the optimal DM.

## Advantages over Moment Method
- Integrates along the dispersion curve → naturally handles noise
- More robust at low S/N
- Less sensitive to scattering tails
